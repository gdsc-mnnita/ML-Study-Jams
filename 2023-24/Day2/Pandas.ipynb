```python
import pandas as pd
data = [[2,4], [4,53], [3,57]]
#row wise is inserted
obj4 = pd.DataFrame(data, index = ['r1', 'r2', 'r3'],columns = ['c1', 'c2'])
print(obj4, '\n')
print('using shape property obj4.shape =',obj4.shape, '\n') # its a tuple
# obj.count() or obj.count(0) or obj.count(axis = 'index')
#                obj.count(1) or obj.count(axis = 'columns')
print('obj4.count(0) will return number of non-NAN(actual values) in column of DF:')
print(obj4.count(0))

print('\nobj4.count(1) will return number of non-NAN(actual vaules) in rows of DF:')
print(obj4.count(1))

print('\nAccessing single element obj4.c2[2]')
print(obj4.c2[2]) # zero based
# or print(obj4.c2['r3'])

#? print(obj4.c2)
```

```python
print(obj4)
print('\n')

# obj[ [ <col1>, <col2>, <col3>, <col4>,... ] ]

# obj.col1

# using loc will include starting and ending index
# obj4.loc[ <startrow> : <endrow>, <startcol> : <endcol> ]
print(obj4.loc['r1':'r2' ,'c1':'c2'])
print('\n')
# changing value
obj4['c3'] = 1234
# what will happen? obj4['r4'] = 567
print(obj4, '\n')
# obj4[1][1] = 34 will not work
obj4['c4'] = [6,7,8]
print(obj4, '\n')

# Row updation and addition
# obj4.loc[<row name>, : ] = [] # list size has to be same as number of columns
# obj4.loc[<row name>, : ] = scalarValue # same value given to all col of the give row

# Coloumn updation and addition
# obj4['c5'] = [] # list size has to be same as number of row in the DF

# Single value updation
# obj.col['row'] = newValue
```

```python
# obj4.groupby('col')
# all the other coloumns are mapped to the uniques values of col elements
print(obj4, '\n\n', 'Grouping\n')

for x, y in obj4.groupby('c3'):
    print(x)
    print(y)
    print()

#renaming rows and columns
# <DF>.rename(index = {<names dictionary>}, columns = {<names dictionary>}, inplace = False)
print("Changing inplace")
print(obj4.rename(index = {'r1':'R1', 'r2':'R2', 'r3':'R3'}, columns = {'c1':'C1', 'c2':'C2', 'c3':'C4', 'r4':'R4', 'c4': 'C5'}, inplace = False))
print("\nNo in place change")
print(obj4)
```

```python
import pandas as pd
r1 = [1,2,3]
r2 = [4,5,6]
r3 = [7,8,9]
r4 = [1000, 2000]
r5 = [4000, 5000]
r6 = [7000, 8000]
df1 = pd.DataFrame(data = [r1, r2, r3], columns = ['A', 'B', 'C'])
df2 = pd.DataFrame(data = [[10,20,30], [40,50,60], [70,80,90]], columns = ['A', 'B', 'C'])
df3 = pd.DataFrame(data = {'A': [100, 400], 'B':[200, 500], 'C':[300, 600]})
df4 = pd.DataFrame(data = [r4, r5, r6], index = [0,1,2], columns = ['A', 'B'])
print(df1)
print()
print(df2)
print()
print(df3)
print()
print(df4)
print()
print("Adding two data frames")
print(df1 + df2)# df = df1 + df3 or df = df1.add(df2)
print()
print("Subtracting two data frames")
print(df3 - df4)# df = df3 - df4 or df = df3.sub(df4)
print()
print("multiplying two data frames")# df = df3*df2 or df = df3.mul(df2)
print(df3*df2)
print()
print("Dividing two data frames")# df = df1 / df4 or df = df1.div(df4)
print(df1 / df4)
print()
```

```python
# axis = 0, along the row for each column (default behaviour)
# axis = 1, along the col for each row
print(df1)
print("getting the minimum for each column")
print(df1.min())# axis = 0, minimum for each column
print()
print("getting the minimum for each row")
print(df1.min(axis = 1))
print()
```
